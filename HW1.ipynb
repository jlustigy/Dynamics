## Problem 1 - Solving the Harmonic Oscillator Numerically

Impliment a differential equation solver for an arbitrary number of degrees of freedom (but designed for a dynamics problem, i.e. each degree pf freedom corresponds to a position and momentum). We want to test **Euler**, **Leap Frog**, and Runge Kutta algorithms. Design code where user can specify which integrator to use.  
**A)** Make a phase portrait of the harmonic oscillator:  
$ \frac{d^2x}{dt^2} = -x $       (1)  
**B)** Solve equation 1 numerically for 0 < t < 70 using **each** of the methods with step sizes 1.6, 0.8, 0.4, 0.2, 0.1, ....  
**C)** Each integrator has a max step size beyond which it is unstable (i.e. where the computed orbit is no longer *qualitatively* the same as the analytic soln). Roughly locate this stepsize for each integrator.

```python
%pylab inline
import numpy as np
import scipy as sp
import galpy as gp
from galpy.orbit import Orbit
import matplotlib.pyplot as plt
from matplotlib import gridspec
from scipy.integrate import ode
matplotlib.rcParams['font.size'] = 16.0
```

Let's first define some useful functions

```python
def euler(y,h,f):
    return y + h*f

def dvdt(t,x):
    return -x

def dxdt(t,v):
    return v

def f(t,y):
    return np.asarray([dxdt(t,y[1]),dvdt(t,y[0])])

def leapfrog(x0,v0,h,tpoints):
    x = np.zeros(tpoints+1)
    v = np.zeros(tpoints+1)
    t = np.arange(tpoints+1)*h
    x[0] = x0
    v[0] = v0
    for i in range(int(tpoints)):
        vhalf = v[i] + 0.5*h*dvdt(0,x[i])
        x[i+1] = x[i] + h*vhalf
        v[i+1] = vhalf + 0.5*h*dvdt(0,x[i+1])
    return x, v, t
```

```python
# This function solves the harmonic oscillator problem for specified initial conditions, 
# step sizes and choice of integrator. 
def integrate(x0=1.0, v0=0.0, tf=70.0, h=0.1, algor="euler", plot=True, color=False, savefig=False):
    
    from colorline import colorline
    
    tpoints = np.ceil(tf/h)
    x = np.zeros(tpoints+1)
    v = np.zeros(tpoints+1)
    t = np.arange(tpoints+1)*h
    y0 = [x0,v0]
    t0 = 0.0
    
    if algor == "euler":
        x = np.zeros(tpoints+1)
        v = np.zeros(tpoints+1)
        t = np.arange(tpoints+1)*h
        x[0] = x0
        v[0]= v0
        for i in range(len(v[1:])):
            v[i+1] = euler(v[i],h,dvdt(0,x[i]))
            x[i+1] = euler(x[i],h,dxdt(0,v[i]))
        print("Integrator: Euler")
        print("Step Size: h="+str(h))
    elif algor == "leapfrog":
        x, v, t = leapfrog(x0,v0,h,tpoints)
        print("Integrator: Leap Frog")
        print("Step Size: h="+str(h))
    elif algor == "rungekutta":
        x = np.zeros(tpoints+1)
        v = np.zeros(tpoints+1)
        t = np.arange(tpoints+1)*h
        x[0] = x0
        v[0] = v0
        r = ode(f).set_integrator('dop853', method='bfd')
        r.set_initial_value(y0, t0)#.set_f_params(2.0)
        for i in range(int(tpoints)):
            r.integrate(r.t+h)
            x[i+1] = r.y[0]
            v[i+1] = r.y[1]
        print("Integrator: Runge Kutta")
        print("Step Size: h="+str(h))
    else:
        print("Using Euler Method as default.")
        
    
    if plot:
        fig = plt.figure(figsize=(10,5))
        gs = gridspec.GridSpec(1, 2)
        ax0 = plt.subplot(gs[0])
        ax1 = plt.subplot(gs[1])
        ax0.plot(t,x)
        ax0.set_xlabel('time')
        ax0.set_ylabel('position')
        ax1.plot(x,v)
        ax1.set_xlabel('position')
        ax1.set_ylabel('velocity')
        plt.tight_layout()
    
    if color:
        fig = plt.figure(figsize=(10,7))
        e = t#v3*v3
        c = colorline(x,v,e, cmap=plt.get_cmap('winter'), norm=plt.Normalize(e.min(), e.max()))
        plt.axis([-1.1,1.1,-1.1,1.1])
        plt.xlabel("position")
        plt.ylabel("velocity")
        cb = plt.colorbar(mappable=c)
        cb.set_label('time', rotation=270, labelpad=20)
        plt.show()
    
    if savefig:
        fig.savefig("hw0_plot_"+algor+"_"+str(h)+".pdf")
```

## Problem 2 - Axisymmetric Potential

Using the $\texttt{galpy}$ package for galactic dynamics, we now want to investigate simple models for the gravitational potential of a galactic disk. Here we begin with the *Two Power Spherical Potential*:  
$$ \rho(r) = \frac{\text{Amp}}{2 \pi a^3} \frac{1}{(r/a)^\alpha + (1+r/a)^{\beta - \alpha}}$$  
If we choose initial conditions:  
> $ R = 1.5$  
$ v_R = 0.1$  
$ v_T = 0.1$  
$ z = 0.1$  
$ v_z = 0.0$  

we get an orbit that varies radially between about .5 and 1.5. We can also see that the orbits precess. 

```python
from galpy.orbit import Orbit
import matplotlib.pyplot as plt
tpsp = gp.potential.TwoPowerSphericalPotential()
#Orbit Initial Conditions: vxvv = [R,vR,vT,z,vz]
o = Orbit(vxvv=[1.5,0.0,0.1,0.1,.0])
ts = numpy.linspace(0,1000,10000)
#ts = numpy.linspace(0,1000,10000)
o.integrate(ts,tpsp,method='odeint')
o.plot3d()
o.plot()
print("Two Power Spherical Potential")
```

Now let's contrast this with the standard *Kepler Potential* ($U(r) = -1/r$) shown below.    

```python
kp = gp.potential.KeplerPotential()
#Orbit Initial Conditions: vxvv = [R,vR,vT,z,vz]
o2 = Orbit(vxvv=[1.5,0.0,.5,0.1,.0])
ts = numpy.linspace(0,100,10000)
o2.integrate(ts,kp,method='odeint')
o2.plot3d()
o2.plot()
print("Kepler Potential")
```

## Problem 3 - Logarithmic Halo Potential

Last but not least, we'll take a look at the non-axisymmetric *Logarithmic Halo Potential*:  
$$ U(x,y) = \frac{1}{2} v_0^2 \text{log} \left ( r_c^2 + x^2 + \frac{y^2}{q^2} \right ) $$  
where  
> $v_0^2 = $ amplitude ($\texttt{amp}$)  
$x = $ core radius at which the logarithm is cut ($\texttt{core}$)  
$q = $ potential flattening ($\texttt{q}$)

#### A. Box Orbits (small energies)

```python
lp = gp.potential.LogarithmicHaloPotential(amp=1.0, core=1, q=0.9, normalize=False)
# Orbit Initial Conditions: vxvv = vxvv=[R,vR,vT,z,vz,phi]
R = 0.05
vR =  0.1
vT = 0.0
z = 0.1
vz = 0.1
phi = 0.0
o = Orbit(vxvv=[R,vR,vT,z,vz,phi])
ts = numpy.linspace(0,1000,10000)
#ts = numpy.linspace(0,1000,10000)
o.integrate(ts,lp,method='odeint')
#o.plot3d()
o.plot(d1='x',d2='z')#d1='R', d2='vR')
print('Logarithmic Halo Potential') 
plt.axis([-.15,.15,-.15,.15])
```

#### B. Loop Orbits (large energies)

```python
lp = gp.potential.LogarithmicHaloPotential(amp=1.0, core=1, q=0.9, normalize=False)
#Orbit Initial Conditions: vxvv=[R,vR,vT,z,vz,phi]
R = 2.5
vR =  1.5
vT = 0.0
z = 0.1
vz = 0.1
phi = 0.0
o = Orbit(vxvv=[R,vR,vT,z,vz,phi])
ts = numpy.linspace(0,1000,10000)
#ts = numpy.linspace(0,1000,10000)
o.integrate(ts,lp,method='odeint')
o.plot(d1='x', d2='z')
print('Logarithmic Halo Potential') 
```
